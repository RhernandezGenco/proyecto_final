```python
import os
import requests

BASE_URL = (
    "https://transtats.bts.gov/PREZIP/"
    "On_Time_Reporting_Carrier_On_Time_Performance_1987_present_{year}_{month}.zip"
)

OUTPUT_DIR = r"D:\Downloads\dd"

os.makedirs(OUTPUT_DIR, exist_ok=True)

years = range(2015,2020)   
months = range(1,13)       

for year in years:
    for month in months:
        url = BASE_URL.format(year=year, month=month)
        filename = os.path.join(
            OUTPUT_DIR,
            f"On_Time_Reporting_Carrier_On_Time_Performance_1987_present_{year}_{month}.zip",
        )

        if os.path.exists(filename):
            print(f"[SKIP] Ya existe: {filename}")
            continue

        print(f"[GET ] {url}")
        resp = requests.get(url, stream=True)

        if resp.status_code == 200:
            total = 0
            with open(filename, "wb") as f:
                for chunk in resp.iter_content(chunk_size=1024 * 1024):
                    if chunk:
                        f.write(chunk)
                        total += len(chunk)
            print(f"[OK  ] Guardado {filename} ({total/1024/1024:.1f} MB)")
        else:
            print(f"[ERR ] {year}-{month:02d} -> HTTP {resp.status_code}")
```

```python
import zipfile
from pathlib import Path

import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

ZIP_DIR = Path(r"D:\Downloads\dd")

OUT_PATH = Path(r"D:\Downloads\dd.parquet")

COLS = [
    "Year", "Month", "DayofMonth", "DayOfWeek", "FlightDate",
    "Reporting_Airline",
    "OriginAirportID", "Origin", "OriginCityName", "OriginState", "OriginWac",
    "DestAirportID", "Dest", "DestCityName", "DestState", "DestWac",
    "CRSDepTime", "DepTime", "DepDelayMinutes", "DepDel15", "DepTimeBlk", "TaxiOut",
    "CRSArrTime", "ArrTime", "ArrDelayMinutes", "ArrDel15", "ArrTimeBlk", "TaxiIn",
    "CRSElapsedTime", "ActualElapsedTime", "Flights", "Distance",
    "Cancelled", "CancellationCode", "Diverted",
    "CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay",
]

def extraer_anio(zpath: Path) -> int | None:
    parts = zpath.stem.split("_")
    try:
        year = int(parts[-2])
        return year
    except Exception:
        return None

zip_files = sorted(ZIP_DIR.glob("On_Time_Reporting_Carrier_On_Time_Performance_1987_present_*.zip"))

zip_files = [z for z in zip_files if 2009 <= (extraer_anio(z) or 0) <= 2019]

print(f"Encontrados {len(zip_files)} ZIP para 2009–2019")

writer = None

for zpath in zip_files:
    print(f"\nLeyendo {zpath.name} ...")

    with zipfile.ZipFile(zpath, "r") as z:
        csv_name = [n for n in z.namelist() if n.lower().endswith(".csv")][0]
        with z.open(csv_name) as f:
            for chunk in pd.read_csv(
                f,
                usecols=lambda c: c in COLS,
                chunksize=200_000,  
                low_memory=False,
            ):
                for c in ["Year", "Month", "DayofMonth", "DayOfWeek",
                          "OriginAirportID", "DestAirportID"]:
                    if c in chunk.columns:
                        chunk[c] = pd.to_numeric(chunk[c], downcast="integer")

                for c in ["DepDelayMinutes", "ArrDelayMinutes",
                          "CRSElapsedTime", "ActualElapsedTime",
                          "Distance",
                          "CarrierDelay", "WeatherDelay", "NASDelay",
                          "SecurityDelay", "LateAircraftDelay"]:
                    if c in chunk.columns:
                        chunk[c] = pd.to_numeric(chunk[c], downcast="float")

                for c in [
                    "Reporting_Airline",
                    "Origin", "OriginCityName", "OriginState", "OriginWac",
                    "Dest", "DestCityName", "DestState", "DestWac",
                    "DepTimeBlk", "ArrTimeBlk", "CancellationCode",
                ]:
                    if c in chunk.columns:
                        chunk[c] = chunk[c].astype("category")

                table = pa.Table.from_pandas(chunk, preserve_index=False)

                if writer is None:
                    writer = pq.ParquetWriter(
                        OUT_PATH,
                        table.schema,
                        compression="snappy",  
                    )
                    print(f"Creando archivo Parquet: {OUT_PATH}")

                writer.write_table(table)

if writer is not None:
    writer.close()
    print("\n=== Listo. Parquet único generado en ===")
    print(OUT_PATH)
else:
    print("\nNo se escribió nada (¿no encontró ZIPs 2009–2019?).")
```

```python
import zipfile
from pathlib import Path

import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

ZIP_DIR = Path(r"D:\Downloads\dd")

OUT_PATH = Path(r"D:\Downloads\dd.parquet")

COLS = [
    "Year", "Month", "DayofMonth", "DayOfWeek", "FlightDate",
    
    "Reporting_Airline",
    
    "OriginAirportID", "Origin", "OriginCityName", "OriginState", "OriginWac",
    
    "DestAirportID", "Dest", "DestCityName", "DestState", "DestWac",
    
    "CRSDepTime", "DepTime", "DepDelayMinutes", "DepDel15", "DepTimeBlk", "TaxiOut",
    
    "CRSArrTime", "ArrTime", "ArrDelayMinutes", "ArrDel15", "ArrTimeBlk", "TaxiIn",
    
    "CRSElapsedTime", "ActualElapsedTime", "Flights", "Distance",
    
    "Cancelled", "CancellationCode", "Diverted",
    
    "CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay",
]

def extraer_anio(zpath: Path):
    parts = zpath.stem.split("_")
    try:
        year = int(parts[-2])
        return year
    except Exception:
        return None

zip_files = sorted(
    z for z in ZIP_DIR.glob("On_Time_Reporting_Carrier_On_Time_Performance_1987_present_*.zip")
    if 2009 <= (extraer_anio(z) or 0) <= 2019
)

print(f"Encontrados {len(zip_files)} ZIP para 2009–2019")

writer = None
schema = None

for zpath in zip_files:
    print(f"\nLeyendo {zpath.name} ...")

    with zipfile.ZipFile(zpath, "r") as z:
        csv_name = [n for n in z.namelist() if n.lower().endswith(".csv")][0]
        with z.open(csv_name) as f:
            for chunk in pd.read_csv(
                f,
                chunksize=200_000,   
                low_memory=False,
            ):
                cols_presentes = [c for c in chunk.columns if c in COLS]
                chunk = chunk[cols_presentes]

                for col in COLS:
                    if col not in chunk.columns:
                        chunk[col] = pd.NA

                chunk = chunk[COLS]

                if schema is None:
                    table = pa.Table.from_pandas(chunk, preserve_index=False)
                    schema = table.schema
                    writer = pq.ParquetWriter(
                        OUT_PATH,
                        schema,
                        compression="snappy",
                    )
                    print(f"Creando archivo Parquet: {OUT_PATH}")
                else:
                    table = pa.Table.from_pandas(
                        chunk,
                        schema=schema,          
                        preserve_index=False,
                    )

                writer.write_table(table)

if writer is not None:
    writer.close()
    print("\n=== Listo. Parquet único generado en ===")
    print(OUT_PATH)
else:
    print("\nNo se escribió nada (¿no encontró ZIPs 2009–2019?).")
```
